Ordenação dos elementos de uma lista pelo método bolha

```python
import matplotlib.pyplot
import random
import time

def bubbleSort(lista):
    for numAtual in range(len(lista)-1,0,-1):
        for i in range(numAtual):
            if lista[i] > lista[i+1]:
                aux = lista[i]
                lista[i] = lista[i+1]
                lista[i+1] = aux


listaTemp = []
tamanho = [500, 1000, 2000, 5000, 10000, 50000, 100000]

for t in tamanho:
    lista = []

    for c in range(t):
        lista.append(random.randint(0,10000))


    tempoInicial = time.time()
    bubbleSort(lista)
    tempoFinal = time.time()
    tempoDuracao = tempoFinal - tempoInicial
    listaTemp.append(tempoDuracao)


matplotlib.pyplot.plot(tamanho, listaTemp)
matplotlib.pyplot.title('Gráfico Método Bolha de Ordenação')
matplotlib.pyplot.xlabel('Dimensão do Vetor')
matplotlib.pyplot.ylabel('Tempo de Ordenação (s)')
matplotlib.pyplot.show()

```
