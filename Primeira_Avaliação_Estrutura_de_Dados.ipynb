#Projeto Análise do Curso de Ciência de Dados:

Código que analisa as respostas dos alunos do curso de CD:

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

df = pd.read_excel('/content/dados_prova.xlsx', index_col=0)


lista = df.values.tolist()
curso = lista[0][2]
print(curso)

pdf = PdfPages('estrutura_de_dados.pdf')
pdf.close()

cs=0
cqs=0
cr=0
cn=0
cnr=0

i=0
j=2

while lista[i][7] == 'ESTRUTURA DE DADOS':

  if lista[i][8] == 'Sempre':
    cs += 1

  elif lista[i][8] == 'Quase Sempre':
    cqs += 1

  elif lista[i][8] == 'Raramente':
    cr += 1

  elif lista[i][8] == 'Nunca':
    cn += 1

  elif lista[i][8] == 'Não respondeu':
    cnr += 1

  i += 1

print(f'{cs} responderam SEMPRE.')
print(f'{cnr} responderam NÃO RESPONDEU')
print(f'{cqs} responderam QUASE SEMPRE.')

resposta = ['SEMPRE', 'NãO RESPONDEU', 'QUASE SEMPRE'] 
  
data = [cs, cnr, cqs]

fig = plt.figure(figsize =(20, 14)) 
plt.pie(data, labels = resposta, autopct='%1.1f%%') 
plt.title('Respostas da Análise do Curso de Ciência de Dados')
plt.savefig('Gráfico_Análise_das_Respostas_do_CD.pdf')
plt.show() 



```
