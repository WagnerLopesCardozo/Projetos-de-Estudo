Exercício 01 - Código que através da definição de uma função lê duas matrizes A e B e efetua a soma dessas matrizes. As dimensões da matriz são definidas pelo usuário

```python
print('*'*109)
print('Programa que lê duas matrizes e efetua a soma entre elas com as dimenções das matrizes definida pelo usuário.')
print('*'*109)

#solicitando o número de linha e colunas da matriz para o usuário

def mostrar_matiz(M):
  for i in M:
    print(i)

nLinhas = int(input('Digite o número de linhas: '))
print()
nColunas = int(input('Digite o número de colunas: '))
print()

M = []

for i in range(nLinhas):
  linha = []

  for j in range(nColunas):
    elemento = input(f'Digite o elemento da linha {i} e o elemento da coluna {j}: ')
    linha.append(int(elemento))
  M.append(linha)
print()
print('-='*38)
print('A matriz digitada pelo usuário e a soma dessas matrizes são respectivamente: ')
print()
print(M)

def soma(A, B):

  M = []

  #temos que verificar se as duas matrizes tem a mesma orde.

  numeroLinhasA, numeroLinhasB = len(A), len(B)
  numeroColunasA, numeroColunasB = len(A[0]), len(B[0])

  if numeroLinhasA == numeroLinhasB and numeroColunasA == numeroColunasB: #satisfazendo esta condição então pode ser feito a soma.

    for i in range(numeroLinhasA):
      linha = [0]*numeroColunasA
      M.append(linha)

      for j in range(numeroColunasA):
        M[i][j] = A[i][j] + B[i][j]

  else:
    print('As matrizes não tem a mesma ordem.')

  return M

print()

print(soma(M, M),end=' ')  

print()
print()

print('FIM DO PROGRAMA')
```

Exercício 01 - Outra forma de se fazer o código da soma de duas matrizes A e B recebendo a dimenção pelo usuário, depois de discutir lógica de programação com colegas.

```python
print('*'*89)
print('Programa que calcula a soma de duas matrizes onde o usuário informa a dimenção da matriz: ')
print('*'*89)

import random

dimensao = int(input('Digite uma dimensão: '))
print()
matrizA = []
matrizB = []
matrizSoma = []

for i in range(0, dimensao):

        linhaA = []
        linhaB = []
        linhaSoma = []

        for j in range(0, dimensao):

            linhaA.append(random.randint(-20, 20))
            linhaB.append(random.randint(-10, 10))
            linhaSoma.append(0)

        matrizA.append(linhaA)
        matrizB.append(linhaB)
        matrizSoma.append(linhaSoma)



for x in range(0, dimensao):
    for y in range(0, dimensao):

        matrizSoma[x][y] = matrizA[x][y] + matrizB[x][y]

def exibirMatriz(matriz):

    for x in range(0, len(matriz)):
        for y in range(0, len(matriz)):

            print(matriz[x][y], end=' ')

        print()

print('A matriz A é: ')

print()

exibirMatriz(matrizA)

print()

print('A matriz B é: ')

print()

exibirMatriz(matrizB)

print()

print('A soma da matriz A com a matriz B será: ')

print()

exibirMatriz(matrizSoma)

print()

print('FIM DO PROGRAMA')
```
